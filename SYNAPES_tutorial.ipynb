# Deep Q Learning

```python
import gym
from agents.DeepQLearning import DQNAgent

# Initialize the CartPole environment and agent
env = gym.make('CartPole-v1')
state_size = env.observation_space.shape[0]
action_size = env.action_space.n
agent = DQNAgent(state_size, action_size)
result = agent.train(env, episodes=1000)
```

# Deep Deterministic Policy Gradient (DDPG)

```python
import gym
from agents.DDPG import DDPGAgent

# Initialize the Pendulum/MountainCar environment and agent
# env = gym.make('Pendulum-v1', g=9.81)
env = gym.make('MountainCarContinuous-v0')

state_size = env.observation_space.shape[0]
action_size = env.action_space.shape[0]
action_max = env.action_space.high.item()

agent = DDPGAgent(state_size, action_size, action_max=action_max)
result = agent.train(env, episodes=500)
```
